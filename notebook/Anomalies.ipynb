# Anomaly Detection in Amazon Cloud Service

## Introduction
Amazon web services is utilized by a large amount of people. Sometimes malicious attackers or curious individuals target these services. This can overload the system architecture. In AWS dataset, CPU utilization, disk writes, and network counts are measured. Due to the sparcity of certain time frames, I located the most attributes that are measured at the same exact time: CPU, ELB and network. This allows a merge without a high number of NaN entries. On a side note, ELB is elastic load balance measurement. This assists with over load of the current server.

Based on these three attributes, do spikes in each attribute show anomalous behavior?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing as prep

from mpl_toolkits.mplot3d import Axes3D #for 3D plots

from sklearn.ensemble import IsolationForest as isoForest
from sklearn.neighbors import LocalOutlierFactor as LOF

#Import datasets with same time sample
CPU=pd.read_csv('../data/external/ec2_cpu_utilization_825cc2.csv')
ELB=pd.read_csv('../data/external/elb_request_count_8c0756.csv')
Network=pd.read_csv('../data/external/ec2_network_in_257a54.csv')
```

## Transforming Data

Based on Google facets, each data set does not contain empty values. Due to this, all three sets are joined together.

Due to the sporadic and high variety of values, each column is normalized on a scale of 0 to 1 in preparation for machine learning algorithms. This ensures consistency.

Lastly, the data is output and saved into a file.

```python
#Alter column names to avoid joining values
CPU.columns=['timestamp','CPU']
ELB.columns=['timestamp','ELB']
Network.columns=['timestamp','Network']

# Merge
Result=CPU.merge(ELB, left_on='timestamp', right_on='timestamp', how='inner')
Result=Result.merge(Network, left_on='timestamp', right_on='timestamp', how='inner')

# ALter types
Result.CPU=Result.CPU.astype('int')
Result.ELB=Result.ELB.astype('int')
Result.Network=Result.Network.astype('int')

#Normalize values from 0-1
mini=Result['CPU'].min()
maxi=Result['CPU'].max()
Result['CPU']=Result['CPU'].apply(lambda x: (x-mini)/(maxi-mini))

mini=Result['ELB'].min()
maxi=Result['ELB'].max()
Result['ELB']=Result['ELB'].apply(lambda x: (x-mini)/(maxi-mini))

mini=Result['Network'].min()
maxi=Result['Network'].max()
Result['Network']=Result['Network'].apply(lambda x: (x-mini)/(maxi-mini))

#Output and save into a file
Result.to_csv('../data/internal/MarchAnalytics.csv')
```

##  Modeling
I surveryed two anomaly detection models: LOF and Isolation Forests.

LOF is calculated on the three attributes: CPU, ELB and Networking. By plotting each attribute against one another with the calculated score, one can see multiple outliers.

```python
Results=pd.read_csv('../data/internal/MarchAnalytics.csv')
LOFData=Results.iloc[:,2:5]

#Convert to numpy array
LOFData=np.array(LOFData)
timestamps=np.array(Results.timestamp)

clf = LOF()
anomalies=clf.fit_predict(LOFData)
Anomal_scores = clf.negative_outlier_factor_

# 3D Scatter Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(LOFData[:, 0], LOFData[:, 1], LOFData[:, 2])
ax.set_xlabel('CPU')
ax.set_ylabel('ELB')
ax.set_zlabel('Networking')
plt.show()
```

#### Isolation Forests

Isolation forests return -1 for anomalies and 1 for normal data. Isolationforest is used as default setup. Since this data is not labelled, the altering of parameters is for discovery use. In the 3D plot, the purple dots represent the anomalies and the yellow dots represent expected values. This is further shown for each attribute in 2D plots.

```python
# fit the model
clf = isoForest()
clf.fit(LOFData)
anomalies = clf.predict(LOFData)


#Start plotting
# 3D Scatter Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(LOFData[:, 0], LOFData[:, 1], LOFData[:, 2],c=anomalies)

ax.set_xlabel('CPU')
ax.set_ylabel('ELB')
ax.set_zlabel('Networking')
plt.show()

plt.title("IsolationForest: CPU Anomalies")
b1 = plt.scatter(LOFData[:, 0], anomalies, c='white', s=20, edgecolor='k')
plt.axis('tight')
plt.xlim((0, 1))
plt.ylim((-1, 1))
plt.show()
plt.title("IsolationForest: ELB Anomalies")
b2 = plt.scatter(LOFData[:, 1], anomalies, c='blue', s=20, edgecolor='k')
plt.show()
plt.title("IsolationForest: Network Anomalies")
b3 = plt.scatter(LOFData[:, 2], anomalies, c='red', s=20, edgecolor='k')
plt.show()
```
